````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

#1 Review your code from the previous exercise and identify areas where you can apply comprehension to simplify and streamline your code. 
def initialize_inventory(products):
    inventory = {product:int(input(f"Enter quantity of {product}: ")) for product in products}
    return inventory

# Initialize the inventory and store it in a global variable
inventory = initialize_inventory(products)
print("Inventory:", inventory)

# Function to get the customer orders
def get_customer_orders():
    
    #2 Prompts the user to enter the number of customer orders
    number_orders = int(input("Enter the number of products you want to order: "))
    
    # Modify the function so it gathers the product names using a loop and user input
    if number_orders > 0:
        # Apply comprehension
        customer_orders = [input("Enter the name of the product you want to order: ") for i in range(number_orders)]
    else:
        print("Invalid input. Please enter a number greater than 0.")
    return customer_orders      

# Call the function to get customer orders
customer_orders = get_customer_orders()
print("Customer's order:", customer_orders)

#3 Function to calculate the total price of the customer order
def total_price(customer_orders):
    prices = [float(input(f"Enter the price of the {product}: ")) for product in customer_orders]
    return sum(prices)

#5 Print the total price
total_price = total_price(customer_orders)
print("Total price:", total_price)

#4 Modify the update_inventory function to remove the product from the inventory if its quantity becomes zero
def update_inventory(customer_orders, inventory):
    updated_inventory = {key: (value - 1 if key in customer_orders and value > 0 else value)for key,value in inventory.items()}

    return updated_inventory

# Call the function 
updated_inventory = update_inventory(customer_orders, inventory)

# Funtion to calculate statistics
def calculate_order_statistics(customer_orders, products):
    unique_products_ordered = len(set(customer_orders))
    percentage_unique_products = (unique_products_ordered / len(products)) * 100
    return unique_products_ordered, percentage_unique_products

# Call the function
order_statistics = calculate_order_statistics(customer_orders, products)

# Function to print order statistics
def print_order_statistics(order_statistics):
    unique_products_ordered, percentage_unique_products = order_statistics
    print(f"Total products ordered: {unique_products_ordered}")
    print(f"Percentage of unique products ordered: {percentage_unique_products:}%")

# Call the function
print_order_statistics(order_statistics)

# Function to print updated inventory
def print_updated_inventory(updated_inventory):
    print("Updated inventory:", updated_inventory)

# Call the function
print_updated_inventory(updated_inventory)
````
